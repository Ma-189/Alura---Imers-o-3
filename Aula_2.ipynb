```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10, 8))

ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order = dados['composto'].value_counts().index[0:5])
ax.set_title('Top 5 Compostos')
plt.show()
```

###**Desafio 02:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'), order = dados['composto'].value_counts().index[0:5])

plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.xlabel('Composto', fontsize='15')
plt.ylabel('Quantidade do composto', fontsize='15')
ax.set_title('Top 5 Compostos', size = 20)
plt.show()
```

###**Desafio 03:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10,8))
dados['g-0'].hist(bins = 100)

plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.xlabel('g-0', fontsize='15')
plt.ylabel('Quantidade', fontsize='15')
```
